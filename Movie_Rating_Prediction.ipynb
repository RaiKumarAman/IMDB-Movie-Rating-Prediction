#ENCODING

```python
df.groupby("Director").describe()
```

```python
df.groupby("Director").agg({"Rating":"mean"})
```

```python
df.groupby("Director").agg({"Rating":"mean"}).to_dict()["Rating"]
```

```python
director_encode=df.groupby("Director").agg({"Rating":"mean"}).to_dict()["Rating"]
```

```python
actor2_encode=df.groupby("Actor 2").agg({"Rating":"mean"}).to_dict()["Rating"]
```

```python
actor3_encode=df.groupby("Actor 3").agg({"Rating":"mean"}).to_dict()["Rating"]
```

```python
actor1_encode=df.groupby("Actor 1").agg({"Rating":"mean"}).to_dict()["Rating"]
```

```python
genre_encode=df.groupby("Genre").agg({"Rating":"mean"}).to_dict()["Rating"]
```

```python
df["actor1_encoded"]=round(df["Actor 1"].map(actor1_encode), 2)
```

```python
df["actor2_encoded"]=round(df["Actor 2"].map(actor2_encode), 2)
df["actor3_encoded"]=round(df["Actor 3"].map(actor3_encode), 2)
```

```python
df["director_encoded"]=round(df["Director"].map(director_encode), 2)
```

```python
df["genre_encoded"]=round(df["Genre"].map(genre_encode), 2)
```

```python
df.head()
```

```python
df.info()
```

```python
x= df[['Year', 'Duration', 'Votes', 'actor1_encoded', 'actor2_encoded', 'actor3_encoded', 'director_encoded', 'genre_encoded']]
y= df["Rating"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
prediction=rf.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test,prediction)
```

```python
metrics.r2_score(y_test,prediction)
```
